```python
#@ 분양용매물제거

df = df[df['TOP'] <= df['listed_year']]
df[['TOP', 'listed_year']]
```

```python
df['address.postal_code'].head()
```

```python
# postal_code가 결측인 것은 일단 999999로 채우기
df['address.postal_code'].fillna(999999, inplace=True)
df['address.postal_code'] = df['address.postal_code'].astype(int)
```

```python
df['address.postal_code'].head()
```

```python
#@title postal sector 추출
df['postal_sector'] = df['address.postal_code'].astype(str).str[:-4]
df[['address.postal_code', 'postal_sector']].head()
```

```python
df['postal_sector']
```

```python
# 왼쪽에 0을 붙여서 두자리수로 만들기
df['postal_sector2']= [num.zfill(2) for num in df['postal_sector']]
```

```python
df['postal_sector2'].head(11)
```

```python
df.columns
```

```python
df.drop(columns = ['postal_sector'], inplace = True)
```

```python
#@title 리스팅 날짜를 year-month format으로 추출해서 다른 컬럼으로 만들기
# Convert the date column to datetime format
df['listed_date'] = pd.to_datetime(df['listed_date'])
```

```python
# Extract year and month from the datetime column
df['month'] = df['listed_date'].dt.month

# Concatenate year and month into a new column 'year_month' as a string
df['year_month'] = df['listed_year'].astype(str) + '-' + df['month'].astype(str).str.zfill(2)
df[['listed_date', 'year_month']].head()
```

```python
#@title Create a pivot table with the count of rows for each 'year_month'
result = pd.pivot_table(df, index='year_month', values='C0', aggfunc='count')
result
```

```python
#@title Calculate the number of rows for each unique combination of values in the 'year_month' and 'contains_virtual' columns
result = df.groupby(['year_month', 'contains_virtual']).size().reset_index(name='count')
# result.to_excel("/content/drive/MyDrive/PropertyGuru/Sale_preprocessed/condo_sale_no.of_listings_by_month.xlsx")
# result.to_excel("/content/drive/MyDrive/PropertyGuru/Sale_preprocessed/HDB_sale_no.of_listings_by_month.xlsx")
result.to_excel("/content/drive/MyDrive/PropertyGuru/Sale_preprocessed/APT_sale_no.of_listings_by_month.xlsx")
```
